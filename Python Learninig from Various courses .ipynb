### <u>Function Shortcircuiting</u>

```python
def avg(*args):
    count=len(args)
    total=sum(args)
    return count and total/count 
## this is called as short circuiting with 'and' if count is false (args has no value len(args) is 0) 
## when no argumetns are passed , 
## total will not be processed
```

```python
avg(2,3,4,5,6,7)
```

### <u> Resursion in function ( Good Example for recursion)</u>

```python
def factorial(n):
    if n<1 :
        return 1
    else:
        print("calculating factorial for {}".format(n))
        return n * factorial(n-1)
```

```python
factorial(4)
```

```python
cache={}
def factorial(n,*, cache): # * makes passing cache compulsory
    if n<1 :
        return 1
    elif n in cache:
            return(cache[n])
    else:
        print("calculating factorial for {}".format(n))
        result=n * factorial(n-1,cache=cache)
        cache[n]=result
        return result
```

```python
factorial(4,cache=cache)
```

```python
cache
```

```python
factorial(5,cache) # it took the value till 4 from cache and calculated it for 5
```

we can also do this with the function 

```python
def factorial(n,cache={}): # * makes passing cache compulsory
    if n<1 :
        return 1
    elif n in cache:
            return(cache[n])
    else:
        print("calculating factorial for {}".format(n))
        result=n * factorial(n-1,cache=cache)
        cache[n]=result
        return result
```

```python
factorial(6) # cache={} will be declared once when function was created, if you don't pass it, it will use the default one 
```

## Clousure

```python
#in below function definition the value of x is not in scope of inner function
# here x (also called free variable) , x in inner is referring to x in outer \
    # and this is called is Closure inner() incloses free variable x .
    # Lecture 101,102 closure < - Amazing
def outer():
    x="Python"
    def inner():
        print("Hello ", x)
    return inner
   
```

```python
fn=outer()
```

```python
#check free variables of a function 
```

```python
fn.__code__.co_freevars # closure object
```

```python
fn.__closure__ # gives a address of x where "Python" is stored
                # Every time you call a function a new scope is created
```

```python
#Closure where we are modifying the closure variable
def outer():
    count=0
    def inner():
        nonlocal count
        count=count+1
        return count
    return inner
```

```python
x=outer()   # when outer is finished running , its scope is gone , 
            # but python saved the reference to count variable through closure
```

```python
x()
```

```python
hex(id(x())) # value of count is immutable , so everytime we call it , is points to another memory address
```

```python
hex(id(x()))
```

```python
x()
```

```python
x.__code__.co_freevars # Closure
```

```python
#multiple clouser
def outer():
    count=0
    def inner():
        nonlocal count
        count=count+1
        return count
    def inner1():
        nonlocal count
        count=count+1
        return count
    return inner,inner1
```
